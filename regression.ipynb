```python
##整理資料/資料預處理
#pandas
import pandas as pd
from sklearn.datasets import load_boston
boston=load_boston()
df=pd.DataFrame(boston["data"],columns=boston["feature_names"])
df["ans"]=boston["target"]
# df.to_csv("boston.csv",encoding="utf-8",index=False)
df
# df.shape
# boston
```

```python
##切成兩分
##資料整理/預處理: pandas機械學習:numpy 
import numpy as np
#np.array(df["ans"])
# df
x = np.array(df.drop(["ans"], axis=1)) ##除掉垂直欄位ans
y = np.array(df["ans"]) #取得分類欄位ans
```

```python
# from sklearn.model_selection import train_test_split
# train_test_split([1, 2, 3, 4], ["a", "b", "c", "d"], test_size=0.25)
# 大部分人會9:1/8:2切
from sklearn.model_selection import train_test_split
# [90%x, 10%x, 90%y, 10%y]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
# x_train.shape
x_test.shape
# y_train
# y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=4)
reg.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(15, 15))
plot_tree(reg, 
     feature_names=boston["feature_names"],
     filled=True)
```

```python
##驗證模型表現如何
#選擇評估標準(分類) 1.正確率 2. f1 score(imblanced)
from sklearn.metrics import r2_score
pre=reg.predict(x_test)
pre
r2_score(y_test,pre)
```
